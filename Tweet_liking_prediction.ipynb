```python
import pandas as pd
import numpy as np

raw_data = pd.read_csv("C:/Users/jsolorzano/Downloads/tweets.csv")

text_data = raw_data[["Tweet"]]
```

```python
from openai import AzureOpenAI
import os
import json

endpoint = os.getenv("ENDPOINT_URL", "")  
deployment = os.getenv("DEPLOYMENT_NAME", "")  
subscription_key = os.getenv("AZURE_OPENAI_API_KEY", "")  

# Initialize Azure OpenAI Service client with key-based authentication    
cliente = AzureOpenAI(  
    azure_endpoint=endpoint,  
    api_key=subscription_key,  
    api_version="2024-05-01-preview",
)

text_data_list = raw_data["Tweet"].tolist()

emb = cliente.embeddings.create(
    model=deployment,  # Nombre del deployment que configuraste en Azure
    input=text_data_list
)


embeddings = [item.embedding for item in emb.data]
raw_data["embeddings"] = embeddings
```

```python
import numpy as np
import pandas as pd

# Ejemplo: Suponiendo que raw_data ya tiene embeddings como arrays NumPy
raw_data["embeddings"] = raw_data["embeddings"].apply(lambda x: np.array(x))  # Asegurar que sean arrays

# Expandir los embeddings en columnas separadas
embeddings_expanded = raw_data["embeddings"].apply(pd.Series)

# Renombrar las columnas para que sean más claras
embeddings_expanded.columns = [f"embedding_{i}" for i in range(embeddings_expanded.shape[1])]

# Combinar con el DataFrame original sin la columna de embeddings original
raw_data_expanded = pd.concat([raw_data.drop(columns=["embeddings"]), embeddings_expanded], axis=1)

# Mostrar las primeras filas del nuevo DataFrame
```

```python
#import xgboost
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(raw_data_expanded.drop(columns=["Tweet", "like"]), raw_data_expanded["like"], test_size=0.5, random_state=42)
```

```python
import mlflow
import mlflow.sklearn
import xgboost as xgb
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Cargar dataset de ejemplo
from sklearn.datasets import make_classification


# Iniciar MLflow
mlflow.set_experiment("Comparación de Modelos ML")

def train_xgboost():
    """Entrena un modelo XGBoost y lo registra en MLflow"""
    with mlflow.start_run(run_name="XGBoost Model"):
        model = xgb.XGBClassifier(
            objective="binary:logistic",
            eval_metric="logloss",
            use_label_encoder=False,
            max_depth=6,
            learning_rate=0.1,
            n_estimators=300,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42,
        )
        
        model.fit(X_train, y_train)
        preds = model.predict(X_test)
        acc = accuracy_score(y_test, preds)
        print(acc)
        print(classification_report(y_test, preds))

        # Registrar en MLflow
        mlflow.log_params(model.get_params())
        mlflow.log_metric("accuracy", acc)
        mlflow.sklearn.log_model(model, "xgboost_model")

        print(f"XGBoost Accuracy: {acc:.4f}")

def train_random_forest():
    """Entrena un modelo RandomForest y lo registra en MLflow"""
    with mlflow.start_run(run_name="RandomForest Model"):
        model = RandomForestClassifier(n_estimators=200, max_depth=6, random_state=42)
        model.fit(X_train, y_train)
        preds = model.predict(X_test)
        acc = accuracy_score(y_test, preds)
        print(acc)
        print(classification_report(y_test, preds))

        # Registrar en MLflow
        mlflow.log_params(model.get_params())
        mlflow.log_metric("accuracy", acc)
        mlflow.sklearn.log_model(model, "random_forest_model")

        print(f"RandomForest Accuracy: {acc:.4f}")

def train_neural_network():
    """Entrena una Red Neuronal con Keras y la registra en MLflow"""
    with mlflow.start_run(run_name="Neural Network Model"):
        model = keras.Sequential([
            layers.Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
            layers.Dense(32, activation="relu"),
            layers.Dense(1, activation="sigmoid"),
        ])
        
        model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
        
        history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=32, verbose=1)
        
        preds = (model.predict(X_test) > 0.5).astype(int)
        acc = accuracy_score(y_test, preds)
        print(acc)
        print(classification_report(y_test, preds))

        # Registrar en MLflow
        mlflow.log_params({"epochs": 10, "batch_size": 32})
        mlflow.log_metric("accuracy", acc)
        mlflow.keras.log_model(model, "neural_network_model")

        print(f"Neural Network Accuracy: {acc:.4f}")

# Entrenar los modelos
train_xgboost()
train_random_forest()
train_neural_network()
```
